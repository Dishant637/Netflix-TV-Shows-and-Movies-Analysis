### Q1- What are the different category of Film or TV Shows which are listed in given data set ? 

Now if we simply do:

```python
netflix_df.listed_in.unique()
```

So we are able to see that In one Movie or TV Show there were different categories listed in. Therefore in the above answer categories are repeating although it is unique because every index of array is different with respect to previous index but we can see that there are many number of categories which are repeating in the array.

Example - TV Mysteries, International TV Shows,etc...

So let's work on getting different number of categories listed in dataset.

```python
lis=[]
lis
```

```python
def different_categories(dataframe = str()):
    import numpy as np
    import pandas as pd
    a= np.array(dataframe.listed_in)
    lis=[]
    for i in range(dataframe.shape[0]):
        m=a[i]
        lis.append((m.split(',')))
    new = pd.Series(np.concatenate(lis)).unique()
    return pd.Series(new)
```

Above function will return the new series of the different category of Film or Tv Show which are listed in given data set.

```python
#Function Calling 
lit = different_categories(netflix_df)
```

```python
listed_in_df = pd.DataFrame(lit, columns= ['listed_in'])
```

```python
listed_in_df 
```

These are different category of Film or TV Show which are listed in given data set.

As we can see above their can be a TV Show which have more than 9 seasons. So in this sorting process there is some kind of ambiguity or something which we are missing.

Here in sorting the column by descending order it is giving the preference to single digit number 9 over any number, so let's try to put 0 in front of every single digit number so that this issue will get resolved. 

```python
new_tv_shows_df = tv_shows_df.reset_index(inplace = False, drop = True) 
```

The function written below will help us to put zero in front of every single digit number and it will return a list of duration as it is by putting zero in front of every single digit number.

```python
def putting_zero_in_front_of_one_digit_integer(dataframe=str()):
    import numpy as np
    a = np.array(dataframe.duration)
    lis =[]
    for i in range(dataframe.duration.shape[0]):
        m = a[i]
        d = m.split(' ')
        d[0] = int(d[0])
        if d[0] == 1 or d[0] == 2 or d[0] == 3 or d[0] == 4 or d[0] == 5 or d[0] == 6 or d[0] == 7 or d[0] == 8 or d[0] == 9:
            d[0] = '0' + str(d[0])
        else:
            d[0] = str(d[0])
        d = d[0] + ' ' + d[1]
        lis.append(d)
    return lis
```

```python
#Function calling
lis1 = putting_zero_in_front_of_one_digit_integer(new_tv_shows_df)
```

```python
lis1
```

```python
new_tv_shows_df['duration']
```

```python
# Replacing older values of duration with new values having zero in front of single digit number
new_tv_shows_df.loc[: , ('duration')] = lis1
```

```python
netflix_df.cast.unique()
```

Here it's not very much clear that total how many casts are their because in every index of array there are too many names. So let's make a function.

```python
def different_cast(dataframe = str()):
    import numpy as np
    import pandas as pd
    a = np.array(dataframe.cast.dropna())
    lis = []
    for i in range(dataframe.cast.dropna().shape[0]):
        m = a[i]
        lis.append((m.split(',')))
    new = pd.Series(np.concatenate(lis)).unique()
    return pd.Series(new)
```
